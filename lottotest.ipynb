```python
# 로또번호 생성 프로그램
# file_pratice3_LottoMaker2.py
# 프로그램을 실행 하면 창이 뜨고 "번호 생성" 버튼을 누르면 아래 텍스트에 번호가 하나씩 생성된다.
# "번호 저장" 버튼을 누르면 번호가 저장된다.
# "번호 삭제" 버튼을 누르면 번호가 삭제된다.
import tkinter as tk
from tkinter.font import Font
import random as rd

# 윈도우를 연다.
window=tk.Tk()
window.title("로또 번호 생성기")
window.resizable(False,False)

# '번호 생성'버튼을 클릭할 경우 1~45 사이의 로또 번호를 만들고 파일에 차례로 저장한다.
def buttonclicked_Make():
    mylist=[]
    while True:
        num=rd.randint(1,45)
        if num not in mylist:
            mylist.append(num)
        if len(mylist) == 6:
            for i in mylist:
                text.insert(tk.END, format(i,'02'))
                text.insert(tk.END,' ')
            text.insert(tk.END,'\n')
            label['text']='번호가 생성되었습니다'
            break

# '번호 저장'버튼을 누를 경우 생성한 모든 번호를 클립 보드에 저장한다.
def buttonclicked_Save():
    window.clipboard_clear() # 기존에 클립보드에 입력되어 있던 내용을 지운다.
    window.clipboard_append(str(text.get('1.0', tk.END))) # 클립보드에 텍스트 위젯 내용을 추가한다.
    window.update()  # 창이 닫힌 이후에도 클립보드에 내용이 유지되도록 한다.
    label['text'] = '번호가 저장되었습니다.'

# '번호 삭제' 버튼을 누르면 생성한 모든 번호를 지운다.
def buttonclicked_Delete():
    text.delete('1.0', tk.END)
    label['text'] = '번호가 삭제되었습니다.'

# 로또 번호를 만드는 버튼
button1=tk.Button(text="번호 생성",command=buttonclicked_Make,width=10)
button1.grid(row=0,column=0)
myFont=Font(size=20)

# 로또 번호를 저장하는 버튼
button2=tk.Button(text="번호 저장",command=buttonclicked_Save,width=10)
button2.grid(row=0,column=1)

# 로또 번호를 지우는 버튼
button3=tk.Button(text="번호 삭제",command=buttonclicked_Delete,width=10)
button3.grid(row=0,column=2)

# 안내 레이블
label=tk.Label(text='대박을 기원합니다')
label.grid(row=1,column=0,columnspan=3)
count=0

# 생성한 번호를 입력할 텍스트
text=tk.Text(font=myFont,width=15,height=10)
text.grid(row=2,column=0,columnspan=3)

window.mainloop()
```
